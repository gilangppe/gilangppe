```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks SPL yang terkait dengan SPL
A = np.array([[0, 0, 1], [1, -1, 0], [-1, 1, 0]])

# Dapatkan matriks uji (x, y, z) yang akan dipertahankan selama metode Jacobi
B = np.array([1, 0, 0])

# Matriks b
b = np.array([[0], [0], [11]])

# Inisialisasi solusi (x, y, z) dengan nilai nol
x = np.array([[0], [0], [0]])

# Tetapkan epsilon
epsilon = 0.0001

# Implementasi metode Jacobi
for i in range(10000):
    x_new = np.dot(A, x) + B
    if np.sum(np.abs(x_new - x)) < epsilon:
        break
    x = x_new

# Plot hasil dari metode Jacobi
plt.plot(x[0], x[1], 'ro')
plt.title('Jacobi Method')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Implementasi metode Gauss-Seidel
def GaussSeidel(A, b, x, epsilon):
    for i in range(10000):
        x_new = np.zeros_like(x)
        for j in range(3):
            sum_other_terms = sum(A[j, k] * x_new[k] for k in range(3) if k != j)
            x_new[j] = (b[j] - sum_other_terms) / A[j, j]
        if np.sum(np.abs(x_new - x)) < epsilon:
            break
        x = x_new
    return x

x = GaussSeidel(A, b, x, epsilon)

# Plot hasil dari metode Gauss-Seidel
plt.plot(x[0], x[1], 'ro')
plt.title('Gauss-Seidel Method')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
